```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the data
data="netflix1.csv"
df=pd.read_csv(data)
df.head()
```

```python
#converted the column into type datetime
df['date_added']=pd.to_datetime(df['date_added'])
df.dtypes
```

```python
df['Year_added']=df['date_added'].dt.year
df.head()
```

```python
#created a new column for month
df['Month']=df['date_added'].dt.month
df.head()
```

```python
month_mapping={1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',
               12:'December'}
for num,name in month_mapping.items():
    df['Month']=df['Month'].replace(num,name)
df.head()
```

```python
#filtered the dataset only for data after year 2000
df=df[df['release_year']>=2000]
df['release_year'].unique()
```
